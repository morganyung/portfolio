```python
df_desks_meta_deduped = df_desks_meta.drop_duplicates(subset='asin')
df_desks_joined = df_desks.merge(df_desks_meta_deduped, on='asin')

# Convert the "category" column to string type
df_desks_joined['category'] = df_desks_joined['category'].astype(str)



```

```python
df_fridge_meta_deduped = df_fridge_meta.drop_duplicates(subset='asin')
df_fridge_joined = df_fridge.merge(df_fridge_meta_deduped, on='asin')

# Convert the "category" column to string type
df_fridge_joined['category'] = df_fridge_joined['category'].astype(str)
```

```python
df_mowers_meta_deduped = df_mowers_meta.drop_duplicates(subset='asin')
df_mowers_joined = df_mowers.merge(df_mowers_meta_deduped, on='asin')

# Convert the "category" column to string type
df_mowers_joined['category'] = df_mowers_joined['category'].astype(str)

# Filter the DataFrame for rows where the "category" column contains "Mowers"
# df_mowers_joined = df_mowers_joined[df_mowers_joined['category'].str.contains("Mowers")]
```

```python
mowers_average_rating_by_asin = df_mowers.groupby('asin')['overall'].mean()
mowers_review_count_by_asin = df_mowers['asin'].value_counts()

# Combine the results into a new DataFrame
df_mowers_scores = pd.DataFrame({
    'Average Overall Rating': mowers_average_rating_by_asin,
    'Review Count': mowers_review_count_by_asin
}).reset_index()

# Rename the 'asin' column to 'Product ID' for clarity
df_mowers_scores.rename(columns={'index': 'asin'}, inplace=True)
df_mowers_scores['product_category'] = 'mowers'
df_mowers_scores = df_mowers_scores.sort_values('Review Count', ascending = False).reset_index().reset_index().drop(columns="index").rename(columns={"level_0":"product category ranking"})
```

```python
desks_average_rating_by_asin = df_desks.groupby('asin')['overall'].mean()
desks_review_count_by_asin = df_desks['asin'].value_counts()

# Combine the results into a new DataFrame
df_desks_scores = pd.DataFrame({
    'Average Overall Rating': desks_average_rating_by_asin,
    'Review Count': desks_review_count_by_asin
}).reset_index()

# Rename the 'asin' column to 'Product ID' for clarity
df_desks_scores.rename(columns={'index': 'asin'}, inplace=True)
df_desks_scores['product_category'] = 'desks'
df_desks_scores = df_desks_scores.sort_values('Review Count', ascending = False).reset_index().reset_index().drop(columns="index").rename(columns={"level_0":"product category ranking"})

```

```python
fridge_average_rating_by_asin = df_fridge.groupby('asin')['overall'].mean()
fridge_review_count_by_asin = df_fridge['asin'].value_counts()

# Combine the results into a new DataFrame
df_fridge_scores = pd.DataFrame({
    'Average Overall Rating': fridge_average_rating_by_asin,
    'Review Count': fridge_review_count_by_asin
}).reset_index()

# Rename the 'asin' column to 'Product ID' for clarity
df_fridge_scores.rename(columns={'index': 'asin'}, inplace=True)
df_fridge_scores['product_category'] = 'fridge'
df_fridge_scores = df_fridge_scores.sort_values('Review Count', ascending = False).reset_index().reset_index().drop(columns="index").rename(columns={"level_0":"product category ranking"})



```

```python
df_ladder_ranks = df_ladder_ranks.rename(columns = {"ranking":"product category ranking"}).drop(columns=["Unnamed: 0"])
```

```python
import pandas as pd


# Concatenate the datasets into one combined DataFrame
df_scores = pd.concat([df_fridge_scores, df_desks_scores, df_mowers_scores, df_ladder_ranks], ignore_index=True)

# Print the combined DataFrame to verify the result
df_scores
```

```python
df = pd.merge(df, df_scores, on='asin', how='left')
```

```python
df.to_excel("FINAL_OUTPUT/data.xlsx")
```
